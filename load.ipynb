```python
#Import the dataset into a dataframe
import pandas as pd
df = pd.read_csv('Current_Assessment_Roll__2022-2023_.csv')
```

```python
#Find the num of na in the dataset
pd.options.display.max_rows = 999
nan_count = df.isna().sum()
print(nan_count)
```

```python
#Remove the columns that does not provide any meaning based on column description and na values
df_x = df[df.columns[~df.columns.isin(['TRACTCE20','GEOID20_tract','GEOID20_blockgroup','Census Block Group','Bill Number','Print Key','Previous Property Class','Fuel Type','Address','Deed Type Description','Basement Type','BOECKH','CONSINCVAL','Special District Code','SWIS','Used as Code','Description 1','Description 2','Description 3'])]]
```

```python
#Remove the POINT() from the location column
df_x.Location = df_x.Location.replace({"POINT":""}, regex=True)
df_x.Location = df_x.Location.str.replace(r'[^(]*\(|\)[^)]*', '')
```

```python
#Assign the latitude and longitude from location, for better precision
df_x[['Longitude','Latitude']] = df_x['Location'].str.split(' ', 1, expand=True)
```

```python
#Remove the Location column
df_x = df_x[df_x.columns[~df_x.columns.isin(['Location'])]]
```

```python
#Find the number of na again
nan_count = df_x.isna().sum()
print(nan_count)
```

```python
#Sort the column based on sbl in ascending order
df_x = df_x.sort_values(by="SBL",ascending=True)
```

```python
#Find the number of na in rows
print(df_x.isnull().sum(axis=1))
```

```python
#Install psycopg2 to communicate to postgresql
!pip install psycopg2
df_x.columns = df_x.columns.str.lower()
```

```python
#Function to estabilish connection to db
def create_conn(database):
    import psycopg2
    from psycopg2 import Error

    try:
        # Connect to an existing database
        connection = psycopg2.connect(user="postgres",
                                      password="1234",
                                      host="localhost",
                                      port="5433",
                                      database=database)
        connection.autocommit = True
        # Create a cursor to perform database operations
        cursor = connection.cursor()
        return cursor

    except (Exception, Error) as error:
        print("Error while connecting to PostgreSQL", error)
    
```

```python
#Create database
cursor = create_conn("postgres")
cursor.execute("""
    SELECT pg_terminate_backend(pg_stat_activity.pid)
    FROM pg_stat_activity
    WHERE pg_stat_activity.datname = 'assessment';
""")
cursor.execute("DROP DATABASE IF EXISTS assessment;")
cursor.close()

# Create a new database
cursor = create_conn("postgres")
sql = '''CREATE database assessment''';

#Creating a database
cursor.execute(sql)
print("Database created successfully........")

#Closing the connection
cursor.close()
```

```python
#Sql query to create the main table with the entire dataset value as raw.
res = 'CREATE TABLE MAIN('
for x,y in enumerate(df_x.columns):
    col = y.replace(" ", "_")
    col = col.replace("#", "no")
    if df_x[y].dtype == 'int64':
        res+=col + ' '+ 'numeric(16,0)'
    elif df_x[y].dtype == 'float64':
        res+=col + ' '+ 'numeric(16,8)'
    else:
        if y != 'sbl':
            res+=col + ' '+ 'character varying(200)'
        else:
            res+=col + ' '+ 'character varying(200) primary key'
    if len(df_x.columns) != x+1:
        res += ','
    else:
        res += ');'
    df_x.rename(columns = {y:col}, inplace = True)
print(res)
```

```python
#Execute the query
cursor = create_conn("assessment")
sql = res;

#Creating a database
cursor.execute(sql)
print("Table created successfully........")

#Closing the connection
cursor.close()
```
